## DETECTING THE OUTLYERS

.OUTLYERS ARE NOTHIN BUT THEY ARE EXTREAM DEVEATED FROM THE MEAN VALUES  

 .THERE ARE THE THREE WAYS TO DETECT THE OUTLYERS  

1.  BOXPLOT
2.  HISTOGRAM
3.  SCATTERPLOT  

HERE WE USE THE BOXPLOT() IN THIS



```python
for col in df.columns: # HERE WE USE THE FOR LOOP TO PRINT THE BOX PLOT FOR THE ALL COLUMNS
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')# An F STRING  ALLOWS YOU TO EMBED VARIABLES DIRECTLY  INTO A STRING.
    plt.show()
```

TREATING THE OUTLYERS WITH "IQR" METHOD

```python
# HERE WE USE THE IQR METHOD
for col in df.columns:
    Q1=df[col].quantile(0.25) #25%
    Q3=df[col].quantile(0.75) #75%
#Q1= FIRST QUANTILE AND Q3=  THIRD QUANTILE
    IQR=Q3-Q1

    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
 # CLIP IS USED TO CAP THE OUTLYERS VALUES
    df[col]=df[col].clip(lower_bound,upper_bound)
```

CHECKING AGAIN FOR THE OUTLYERS

```python
for col in df.columns:
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

CORRELEATION

```python
corr=df[["LB","AC","FM","UC","DL","ASTV","MSTV","ALTV","MLTV","Width","Tendency"]].corr()
corr

# THE CORR() IS USED TO FIND THE CORRELEATION BETWEEN THE VARIABLES
```

. PLOTING THE DISTPLOT AND THE VIOLINE PLOT

```python
for col in df.columns:
    if col not in(["DS","DP","NSP"]): # NOT IN OPERATOR IS USED, IT WILL NOT SELECT THIS VARIABLES
        plt.figure(figsize=(12,7))
        plt.subplots_adjust(hspace=0.4, wspace=0.4)# hspace IS HORIZANTAL SPACE
                                                   # hspace IS VERTICAL SPACE
        #distplot
        plt.subplot(121)
        sns.distplot(df[col],color="orange")
        plt.title(f'Histogram of {col}')

        #violin plot
        plt.subplot(122)
        sns.violinplot(x=col,data=df,color="yellow")
        plt.title(f'Distribution of {col}')
        plt.show()
```

HISTOGRAM USIN THE MATPLOT

```python

for col in df.columns:
    plt.figure(figsize=(12, 7))  # Create a new figure
    plt.hist(df[col], bins=8, color="orange", edgecolor="black")  # Histogram
    plt.title(f'Histogram of {col}')  # Title
    plt.xlabel(col)  # Label x-axis
    plt.ylabel("Frequency")  # Label y-axis
    plt.show()  # Show the plot
```
